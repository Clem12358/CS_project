```python
import pandas as pd
import numpy as np
```

```python
SwissHoliday0 = pd.read_csv("/content/drive/My Drive/GroupProject_BusinessAnalytics/swissholiday.csv")
SwissHoliday0.head()
```

```python
SwissHoliday0 = SwissHoliday0[['Date', 'Counties']]
SwissHoliday = SwissHoliday0.rename(columns={'Counties': 'Cantons'})
SwissHoliday.head()
```

```python
SwissHoliday
```

```python
SwissHoliday['Cantons'] = SwissHoliday['Cantons'].str.replace('CH-', '')
```

```python
SwissHoliday
```

```python
import pandas as pd

# Assuming you have already loaded your DataFrame as 'SwissHoliday'

# Get all unique cantons from the 'Cantons' column
# Convert the 'Cantons' column to string type before applying str.split
all_cantons = SwissHoliday['Cantons'].astype(str).str.split(',').explode().unique()

# Print the unique cantons
print(all_cantons)

```

```python
import pandas as pd

# Assuming you have already loaded your DataFrame as 'SwissHoliday'

# List of cantons to add as columns
cantons_to_add = ['nan', 'BE', 'FR', 'SH', 'AG', 'TG', 'VD', 'TI', 'UR', 'SZ', 'NE', 'VS', 'LU', 'NW',
                  'GL', 'ZH', 'OW', 'ZG', 'SO', 'BS', 'BL', 'AR', 'AI', 'SG', 'GR', 'GE', 'JU']

# Create indicator columns for each canton
for canton in cantons_to_add:
    SwissHoliday[canton] = SwissHoliday['Cantons'].str.contains(canton, na=False).astype(int)

SwissHoliday.loc[SwissHoliday['Cantons'].isnull(), cantons_to_add] = 1
# Display the updated DataFrame
SwissHoliday.head()
```

```python
SwissHoliday = SwissHoliday[['Date', 'BE', 'VD', 'TI', 'VS', 'LU', 'ZH', 'BS', 'SG', 'GE']]
```

```python
SwissHoliday['Date'] = pd.to_datetime(SwissHoliday['Date'], format='%m-%d-%Y') # Correct format is '%m-%d-%Y'
#SwissHoliday['Date'] = pd.to_datetime(SwissHoliday['Date'], dayfirst=True)

# 2. Format to 'm-day-year' (month-day-year)
SwissHoliday['Date'] = SwissHoliday['Date'].dt.strftime('%m-%d-%Y')
```
